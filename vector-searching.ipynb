```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
from langchain_chroma import Chroma
import os
```

```python
# to store credential we need .env file for secure access

from dotenv import load_dotenv

load_dotenv()
```

```python
import pandas as pd

books = pd.read_csv("cleaned_data.csv")
```

```python
books["tagged_description"]
```

```python
#converting text file to csv
books["tagged_description"].to_csv("tagged_description.txt", index=False, header=False)
```

```python
# Load with UTF-8 encoding
raw_documents = TextLoader("tagged_description.txt", encoding="utf-8").load()

# Now split the text into documents
text_splitter = CharacterTextSplitter(chunk_size=500, chunk_overlap=50, separator="\n")
documents = text_splitter.split_documents(raw_documents)
```

```python
documents[0]
```

```python
#now let's come towards our vector database
```

```python
embedding = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

db_books = Chroma.from_documents(documents, embedding=embedding)

# when we will run the file as admistrator it will not give warnings
```

```python
query = "A book about human psychology"
docs = db_books.similarity_search(query, k = 10)
docs
```

```python
isbn = int(docs[0].page_content.split()[0].strip().strip('"'))
books[books["isbn13"] == isbn]
```

```python
def retrieve_semantic_recommendations(
        query: str,
        top_k: int = 10,
) -> pd.DataFrame:
    recs = db_books.similarity_search(query, k = 50)

    books_list = []

    for i in range(0, len(recs)):
        books_list += [int(recs[i].page_content.strip('"').split()[0])]

    return books[books["isbn13"].isin(books_list)]
```

```python
retrieve_semantic_recommendations("romantic kind of books")
```
